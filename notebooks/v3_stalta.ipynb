```python
from obspy import UTCDateTime
from obspy.clients.fdsn import Client
from obspy.clients.fdsn.header import FDSNNoDataException
from IPython.display import clear_output
import matplotlib.pyplot as plt
import time

from obspy.signal.trigger import classic_sta_lta, trigger_onset
```

```python
NETWORK = "IV"
STATION = "GIZZ"
LOCATION = "*"
CHANNEL = "HHZ"

WINDOW = 60
INTERVAL = 5
```

```python
client = Client("INGV")
```

```python
endtime = UTCDateTime.now() - 20
starttime = endtime - 120
```

```python
st = client.get_waveforms(
    network=NETWORK,
    station=STATION,
    location=LOCATION,
    channel=CHANNEL,
    starttime=starttime,
    endtime=endtime
)
```

```python
tr = st[0].copy()

# preprocessing
tr.detrend("demean")
tr.detrend("linear")
tr.filter("bandpass", freqmin=1.0, freqmax=15.0)
```
